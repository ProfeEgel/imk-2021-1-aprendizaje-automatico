```python
import numpy as np
import matplotlib.pyplot as plt
```

# Ejemplo simplificado $h_\theta(x) = \theta_1 X$

```python
X = np.arange(0,5,1)
y = X
X_cont = np.arange(0,5,0.1)
y_cont = X_cont
```

```python
plt.figure(figsize=(5,5))
plt.xlabel("X", fontsize=12)
plt.ylabel("y", fontsize=12)
plt.plot(X_cont, y_cont, "b-", linewidth=2)
plt.plot(X, y, "ro", markersize=6)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.xlabel("X", fontsize=12)
plt.plot(X, y, "ro", markersize=6)

theta1 = 1.5
h = theta1 * X
plt.plot(X, h, "g--", linewidth=2)

plt.legend(["y", r"$h_\theta(x)$"])
plt.grid(True)
plt.show()
```

```python
m = X.size
theta1 = np.arange(-3,5,0.1)

h = np.array([theta*X for theta in theta1])
J = (1/(2*m))*(((h - y)**2).sum(axis=1))

plt.figure(figsize=(5,5))
plt.xlabel(r"$\theta_1$", fontsize=12)
plt.ylabel(r"$J(\theta_1)$", fontsize=12)
plt.plot(theta1,J, "b-", markersize=6)
plt.grid(True)
plt.show()
```

# Ejemplo $h_\theta(x) = \theta_0 + \theta_1 X$

```python
X = np.arange(-2,2,0.1)
y = 2.5 + 5*X
```

## MAE (Mean Absolute Error)
$$J(\theta) = \dfrac{1}{2m}\sum_{i=1}^m|(h_\theta(x)-y)|$$

## MSE (Mean Squared Error)
$$J(\theta) = \dfrac{1}{2m}\sum_{i=1}^m(h_\theta(x)-y)^2$$

## RMSE (Root Mean Squared Error)
$$J(\theta) = \sqrt{\dfrac{1}{2m}\sum_{i=1}^m(h_\theta(x)-y)^2}$$

```python
m = X.size
theta0 = np.arange(-12.5,17.5,0.1)
theta1 = np.arange(-10,20,0.1)
h = np.array([[th0+th1*X  for th0 in theta0] for th1 in theta1])
J = (1/(2*m))*(((h - y)**2).sum(axis=2))
```

```python
fig = plt.figure()
ax = plt.axes(projection="3d")
th0, th1 = np.meshgrid(theta0, theta1)
surf = ax.plot_surface(th0, th1, J, cmap="coolwarm")

ax.set_xlabel(r"$\theta_0$", fontsize=12)
ax.set_ylabel(r"$\theta_1$", fontsize=12)
ax.set_zlabel(r"$J(\theta_1)$", fontsize=12)
plt.show()

plt.contour(th0, th1, J, 50, cmap="coolwarm")
plt.xlabel(r"$\theta_0$", fontsize=12)
plt.ylabel(r"$\theta_1$", fontsize=12)
plt.colorbar()
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.xlabel("X", fontsize=12)
plt.plot(X, y, "ro", markersize=6)

theta0 = 5
theta1 = 5
h = theta0 + theta1*X
plt.plot(X, h, "g--", linewidth=2)

plt.legend(["y", r"$h_\theta(x)$"])
plt.grid(True)
plt.show()
```
