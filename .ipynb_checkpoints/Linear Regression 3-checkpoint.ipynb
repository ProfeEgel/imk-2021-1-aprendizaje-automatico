```python
X = df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population']]
y = df['Price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=55)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
(model.intercept_, model.coef_)
```

```python
y_predictions = model.predict(X_test)
```

```python
plt.scatter(y_predictions, y_test)
```

```python
from sklearn import metrics
```

```python
print(metrics.mean_absolute_error(y_test, y_predictions))
print(metrics.mean_squared_error(y_test, y_predictions))
print(np.sqrt(metrics.mean_squared_error(y_test, y_predictions)))
```
