```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
z = np.linspace(-10,10,50)
```

```python
plt.plot(z, sigmoid(z), "b-", linewidth=2)
plt.xlabel("z", fontsize=16)
plt.ylabel("g(z)", fontsize=16)
plt.grid(True)
plt.show()
```

```python
X = 0.1*np.array([0.52, 0.81, 0.96, 1.58, 2.42, 2.53, 3.96, 4.27, 4.59, 5.72])
y = np.array([0, 0, 0, 0, 1, 0, 1, 1, 1, 1])
```

```python
plt.plot(X, y, "ro", markersize=6)
plt.xlabel("X", fontsize=12)
plt.ylabel("y", fontsize=12)
plt.grid(True)
plt.show()
```

```python
m = X.size

theta1 = np.arange(-100,100,0.1)
# h = np.array([theta*X for theta in theta1]) # linear regression
h = np.array([sigmoid(theta*X) for theta in theta1]) # logistic regression
J = (1/(2*m))*(((h - y)**2).sum(axis=1))

plt.xlabel(r"$\theta_1$", fontsize=12)
plt.ylabel(r"$J(\theta_1)$", fontsize=12)
plt.plot(theta1,J, "b-", markersize=6)
plt.grid(True)
plt.show()
```

```python
h = np.arange(0.01, 1, 0.01)
J_1 = -np.log(h)
J_0 = -np.log(1-h)
```

```python
plt.plot(h, J_0, "b-", markersize=6, label="y=0")
plt.plot(h, J_1, "r-", markersize=6, label="y=1")
plt.xlabel(r"$h(\theta)$", fontsize=12)
plt.ylabel(r"$J(\theta)$", fontsize=12)
plt.legend(fontsize=12)
plt.grid(True)
plt.show()
```
